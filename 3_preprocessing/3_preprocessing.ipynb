```python
%load_ext autoreload
%autoreload 2

import sys
import os

parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.insert(0, parent_dir)
```

# Getting the data

As always, let's import our last updated data.

```python
from utils import load_datasets

previous_folder = "2_hashtag_enrichment"
gru_train_set, gru_val_set, gru_test_set = load_datasets(folder=os.path.join(parent_dir, previous_folder, "output"))
grubert_train_set, grubert_val_set, grubert_test_set = load_datasets(folder=os.path.join(parent_dir, previous_folder, "output"))
```

Gru Preprocessing

```python
from preprocessing import fundamental_preprocessing, emoji_and_emoticons_preprocessing

gru_train_set = emoji_and_emoticons_preprocessing(data=gru_train_set, parent_dir=parent_dir, verbose=True)
gru_train_set = fundamental_preprocessing(
        data=gru_train_set, 
        lowercase=True, 
        remove_urls=True, 
        remove_video_placeholders=True,
        remove_mentions=True, 
        remove_html_entities=True, 
        remove_retweet_markers=True, 
        replace_abbreviations=True, 
        remove_duplicates_at_end=True,
        keep_only_good_symbols=True,
        normalize_whitespace=True, 
        verbose=True
     )

gru_val_set = emoji_and_emoticons_preprocessing(data=gru_val_set, parent_dir=parent_dir, verbose=True)
gru_val_set = fundamental_preprocessing(
        data=gru_val_set, 
        lowercase=True, 
        remove_urls=True, 
        remove_video_placeholders=True,
        remove_mentions=True, 
        remove_html_entities=True, 
        remove_retweet_markers=True, 
        replace_abbreviations=True, 
        remove_duplicates_at_end=True,
        keep_only_good_symbols=True,
        normalize_whitespace=True, 
        verbose=True
     )

gru_test_set = emoji_and_emoticons_preprocessing(data=gru_test_set, parent_dir=parent_dir, verbose=True)
gru_test_set = fundamental_preprocessing(
        data=gru_test_set, 
        lowercase=True, 
        remove_urls=True, 
        remove_video_placeholders=True,
        remove_mentions=True, 
        remove_html_entities=True, 
        remove_retweet_markers=True, 
        replace_abbreviations=True, 
        remove_duplicates_at_end=True,
        keep_only_good_symbols=True,
        normalize_whitespace=True, 
        verbose=True
     )

gru_output_folder = os.path.join(parent_dir, "3_preprocessing", "output", "gru")
os.makedirs(gru_output_folder, exist_ok=True)

for subset, name in zip([gru_train_set, gru_val_set, gru_test_set], ["train", "val", "test"]):
    subset.to_csv(os.path.join(gru_output_folder, f"{name}.csv"), index=False)
```
